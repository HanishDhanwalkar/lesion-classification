```python
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig

import torch
torch.manual_seed(1234)
```

```python
quantization_config = GPTQConfig(bits=4, disable_exllama=True) # Disable exllama to load on CPU

model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen-VL-Chat-Int4",
    quantization_config=quantization_config, # Pass quantization config
    device_map="cuda:0",
    trust_remote_code=True
).eval()
```

```python
# Note: The default behavior now has injection attack prevention off.
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen-VL-Chat", trust_remote_code=True)
```

```python
# 1st dialogue turn
query = tokenizer.from_list_format([
    {'image': 'https://qianwen-res.oss-cn-beijing.aliyuncs.com/Qwen-VL/assets/demo.jpeg'},
    {'text': 'What is this?'},
])
response, history = model.chat(tokenizer, query=query, history=None)
print(response)
# Output: This is an image featuring a woman sitting on a beach sand playing with a dog.
```

```python
# 2nd dialogue turn
response, history = model.chat(tokenizer, 'Output the detection box of "high five"', history=history)
print(response)
# <ref>"high five"</ref><box>(517,508),(589,611)</box>

image = tokenizer.draw_bbox_on_latest_picture(response, history)
if image:
  image.save('1.jpg')
else:
  print("no box")
```

```python
### Image
import os

imgs = os.listdir()
imgs = [img for img in imgs if img.endswith('.jpg')]
imgs
```

```python
imgs[0]
```

```python
query = tokenizer.from_list_format([
    {'image': imgs[0]},
    {'text': "This is a close-up of a person's mouth. He/she may be affected with lesions inside of the mouth. The possible types of lesion are: 1)leukoplakia, 2)tumor, 3)ulcer, 4)erythroleukoplakia. Find the type of lesion. Return only the type of lesion and nothing else."},
])
response, history = model.chat(tokenizer, query=query, history=None)
print("1st round: ",response)

```

```python
lesion_type = response.split('\n')[0]
lesion_type
```

```python
response, history = model.chat(tokenizer, f'Output the detection box of {lesion_type}. The box should be as small as possible. Try to exclude teeth from the box.', history=history)
print(response)
# <ref>"...."</ref><box>(517,508),(589,611)</box>

image = tokenizer.draw_bbox_on_latest_picture(response, history)
if image:
  image.save(f'sample_result_{lesion_type}.jpg')
else:
  print("no box")
```
